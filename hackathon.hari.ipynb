```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df2 = scaler.fit_transform(df1)
```

```python
scaled_data = pd.DataFrame(scaled_df2)
```

```python
scaled_data.head()
```

```python
from sklearn.cluster import KMeans
kmean =KMeans(n_clusters=2)
```

```python
kmean.fit(scaled_data)
```

```python
pred = kmean.predict(scaled_data)
```

```python
pred
```

```python
pd.DataFrame(pred).value_counts()
```

```python
kmean.score(scaled_data)
```

```python
kmean.inertia_
```

```python
SSE = []
```

```python
for cluster in range(1,50):
    kmean = KMeans(n_jobs = -1,n_clusters = cluster)
    kmean.fit(scaled_data)
    SSE.append(kmean.inertia_)
```

```python
plt.figure(figsize=(12,6))
frame_scaled = pd.DataFrame({'cluster':range(1,50),'SSE':SSE})
plt.plot(frame_scaled['cluster'],frame_scaled['SSE'],marker='o')
plt.xlabel('cluster')
plt.ylabel('SSE')
```

```python
kmean = KMeans(n_jobs = -1,n_clusters = 40)
kmean.fit(scaled_data)
```
